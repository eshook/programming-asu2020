# Simple Interactive Examples
This notebook is made available to show you Jupyter Widgets, which allow for interactivity in a Jupyter Notebook. In Computer Science terms they provide user interface (UI) controls for working with code and data interactively. For more information please look at the following websites:

```python
# Here we are going to use overtime pay as an example from 3_Fundamentals.

# We are going to import a number of packages, this is from
# https://ipywidgets.readthedocs.io/en/latest/examples/Using%20Interact.html
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

# Here we have defined a Python function that accepts the number of hours
# someone has worked. Then it will calculate the pay for that person
def calculate_pay(numberofhours):
    payrate = 10.0
    overtimerate = 2.0
    overtimethreshold = 40

    pay = None

    if numberofhours <= overtimethreshold:
        pay = payrate * numberofhours
    else:
        #             Number of overtime hours * our time and a half payrate
        overtimepay = (numberofhours - overtimethreshold) * payrate * overtimerate
        normalpay = overtimethreshold * payrate
        pay = normalpay + overtimepay
    # Once we have pay calculated we will return the value
    # Which will be assigned to the pay variable below
    return pay

# Here we are going to pass in our calculate_pay function
# And the numberofhoursvariable to be 40
# Notice the slider allows for negative hours. That doesn't make sense. 
# So let's improve it in the next code cell
interact(calculate_pay, numberofhours=40)

```

```python
# Here is an example from the Seaborn gallery
# https://seaborn.pydata.org/examples/hexbin_marginals.html
import numpy as np
import seaborn as sns
sns.set_theme(style="ticks")

rs = np.random.RandomState(11)
x = rs.gamma(2, size=1000)
y = -.5 * x + rs.normal(size=1000)

sns.jointplot(x=x, y=y, kind="hex", color="#4CB391")
```

```python
# Here is an example from the Seaborn gallery
# https://seaborn.pydata.org/examples/hexbin_marginals.html
import numpy as np
import seaborn as sns

# Make it a function
def plotfunction(slidegamma):
    sns.set_theme(style="ticks")

    rs = np.random.RandomState(11)
    # Notice we changed 2 to slidegamma
    x = rs.gamma(slidegamma, size=1000)
    y = -.5 * x + rs.normal(size=1000)

    sns.jointplot(x=x, y=y, kind="hex", color="#4CB391")
    
    return sns

# Call the function
plotfunction(8)
```
